```python
import cv2
import numpy as np

def crop_face(image_path, output_path="/content/cropped_face.jpg", crop_size=(256, 256)):
    image = cv2.imread(image_path)
    if image is None:
        print(f"Error: Could not read image at {image_path}")
        return None

    image = cv2.resize(image, (image.shape[1]*2, image.shape[0]*2), interpolation=cv2.INTER_CUBIC)
    h, w, _ = image.shape

    face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.1, 4)

    if len(faces) == 0:
        print("No face detected!")
        return None

    x, y, width, height = faces[0]

    margin = 0.2
    x1 = max(0, int(x - width * margin))
    y1 = max(0, int(y - height * margin))
    x2 = min(w, int(x + width * (1 + margin)))
    y2 = min(h, int(y + height * (1 + margin)))

    face_crop = image[y1:y2, x1:x2]
    face_crop_resized = cv2.resize(face_crop, crop_size)
    cv2.imwrite(output_path, face_crop_resized)

    print(f"Cropped face saved to {output_path}")
    return face_crop_resized, (x1, y1, x2, y2)
```

```python
import cv2
import numpy as np

def blend_animated_face(original_image_path, animated_face_video_path, output_path="/content/blended_video.mp4"):
    original_img = cv2.imread(original_image_path)
    if original_img is None:
        print(f"Could not read image: {original_image_path}")
        return None

    img_height, img_width = original_img.shape[:2]

    video = cv2.VideoCapture(animated_face_video_path)
    if not video.isOpened():
        print(f"Could not open video: {animated_face_video_path}")
        return None

    fps = video.get(cv2.CAP_PROP_FPS)
    frame_count = int(video.get(cv2.CAP_PROP_FRAME_COUNT))

    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    out = cv2.VideoWriter(output_path, fourcc, fps, (img_width, img_height))

    face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
    gray = cv2.cvtColor(original_img, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.1, 4)

    if len(faces) == 0:
        print("No face detected in original image")
        width = img_width // 2
        height = img_height // 2
        x = (img_width - width) // 2
        y = (img_height - height) // 2
    else:
        x, y, width, height = faces[0]
        margin = 0.2
        x = max(0, int(x - width * margin))
        y = max(0, int(y - height * margin * 1.5))  # Add more margin at the top
        width = min(img_width - x, int(width * (1 + 2 * margin)))
        height = min(img_height - y, int(height * (1 + 2 * margin)))

    print(f"Face position: x={x}, y={y}, width={width}, height={height}")

    x = max(0, min(x, img_width - 1))
    y = max(0, min(y, img_height - 1))
    width = max(1, min(width, img_width - x))
    height = max(1, min(height, img_height - y))

    for i in range(frame_count):
        ret, face_frame = video.read()
        if not ret:
            break

        face_resized = cv2.resize(face_frame, (width, height))

        result = original_img.copy()

        mask = np.zeros((height, width), dtype=np.uint8)

        center_x = width // 2
        center_y = height // 2

        ellipse_width = int(width * 0.9)
        ellipse_height = int(height * 0.95)

        cv2.ellipse(
            mask,
            (center_x, center_y),
            (ellipse_width // 2, ellipse_height // 2),
            0, 0, 360, 255, -1
        )

        mask = cv2.GaussianBlur(mask, (51, 51), 30)

        mask_3ch = np.stack([mask] * 3, axis=2) / 255.0

        result[y:y+height, x:x+width] = (
            face_resized * mask_3ch +
            result[y:y+height, x:x+width] * (1 - mask_3ch)
        )

        out.write(result.astype(np.uint8))

    video.release()
    out.release()

    print(f"Blended video saved to {output_path}")
    return output_path
```

```python
import cv2
import numpy as np

def blend_face_position(original_image_path, animated_face_video_path, output_path="/content/blended_video.mp4"):
    original_img = cv2.imread(original_image_path)
    if original_img is None:
        print(f"Could not read image: {original_image_path}")
        return None

    img_height, img_width = original_img.shape[:2]

    video = cv2.VideoCapture(animated_face_video_path)
    if not video.isOpened():
        print(f"Could not open video: {animated_face_video_path}")
        return None

    fps = video.get(cv2.CAP_PROP_FPS)
    frame_count = int(video.get(cv2.CAP_PROP_FRAME_COUNT))

    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    out = cv2.VideoWriter(output_path, fourcc, fps, (img_width, img_height))

    face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
    gray = cv2.cvtColor(original_img, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.1, 4)

    if len(faces) == 0:
        print("No face detected in original image, using center position")
        center_x = img_width // 2
        center_y = img_height // 2
        face_size = min(img_width, img_height) // 3
        x = center_x - face_size // 2
        y = center_y - face_size // 2
        width = face_size
        height = face_size
    else:
        x, y, width, height = faces[0]

    print(f"Face position: x={x}, y={y}, width={width}, height={height}")

    for i in range(frame_count):
        ret, face_frame = video.read()
        if not ret:
            break

        result = original_img.copy()

        try:
            face_resized = cv2.resize(face_frame, (width, height))

            # Create mask for smooth blending
            mask = np.zeros((height, width), dtype=np.uint8)
            cv2.ellipse(mask, (width//2, height//2), (width//2-10, height//2-10), 0, 0, 360, 255, -1)
            mask = cv2.GaussianBlur(mask, (21, 21), 11)
            mask_3d = np.expand_dims(mask, axis=2) / 255.0

            # Extract the region of interest
            roi = result[y:y+height, x:x+width]

            # Blend only where mask is non-zero
            blended = (face_resized * mask_3d) + (roi * (1 - mask_3d))

            # Place the blended result back into the image
            result[y:y+height, x:x+width] = blended

        except Exception as e:
            print(f"Error in frame {i}: {e}")
            # Try with perfect fit
            face_resized = cv2.resize(face_frame, (width, height))
            result[y:y+height, x:x+width] = face_resized

        out.write(result)

    video.release()
    out.release()

    print(f"Blended video saved to {output_path}")
    return output_path
```
